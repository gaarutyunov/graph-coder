```python
from graph_coder.train.train import train, setup_parser
from argparse import ArgumentParser
```

```python
from argparse import Namespace, _StoreAction, _StoreTrueAction

parser = ArgumentParser()
setup_parser(parser)

args = {}

for action in parser._actions:
    if isinstance(action, (_StoreAction, _StoreTrueAction)):
        args[action.dest] = action.default

namespace = Namespace(**args)
namespace.data = "~/git-py"
```

```python
train(namespace)
```
